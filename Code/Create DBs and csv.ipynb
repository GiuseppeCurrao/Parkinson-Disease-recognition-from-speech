# DB cvs creation
In this code we created the csv to access to the .wav files from the other code files. 
The csv file has 3 label: index, path and true label of the file.
THE CODE IS NOT RUN BECOUSE MOST OF THE WORK WAS DONE IN LOCAL USING ANACONDA  


This function returns a list for all the files in all the folders and subfolders. 

```python
def list_files(path):
    files = []
    for r, d, f in os.walk(path):
        for file in f:
            files.append(os.path.join(r, file))
    return files
```

Create the csv file taking the .wav file from the the folders, assigning the label 0 for healthy persons and 1 for PD patients. 

```python
import csv
audio_dir1 = "/content/drive/MyDrive/DB/15 Young Healthy Control"
audio_dir2 = '/content/drive/MyDrive/DB/22 Elderly Healthy Control'
audio_dir3 = '/content/drive/MyDrive/DB/28 People with Parkinson disease'
files1 = list_files(audio_dir1)
files2 = list_files(audio_dir2)
files3 = list_files(audio_dir3)
with open("file.csv", "w", newline="") as file:
    writer = csv.writer(file)
    i = 0
    writer.writerow(['index', 'path', 'label'])
    # Ciclo sui file nella directory
    for filename in files1:
        if filename.endswith(".wav"):
            # Scrivi il path del file nel file CSV
            writer.writerow([i,os.path.join(audio_dir1, filename), 0])
            i+=1
    
    i=0
    for filename in files2:
        if filename.endswith(".wav"):
            # Scrivi il path del file nel file CSV
            writer.writerow([i,os.path.join(audio_dir1, filename), 0])
            i+=1

    i=0
    for filename in files3:
        if filename.endswith(".wav"):
            # Scrivi il path del file nel file CSV
            writer.writerow([i,os.path.join(audio_dir1, filename), 1])
            i+=1
```

# Creation of the segmented database
In the CNN models we do not use the entire audio spectrogram, but in section of 8 ms. So, starting from the entire files, we create another DB and csv.

```python
import pandas as pd
!pip install pydub
from pydub import AudioSegment
import os

# Function to create segment of audio files
def split_wav_with_labels(file_path, label, output_folder, index, duration):
    # Carica il file WAV
    audio = AudioSegment.from_wav(file_path)
    
    segment_duration = duration * 1000
    
    for i, start_time in enumerate(range(0, len(audio), segment_duration)):
        
        segment = audio[start_time:start_time + segment_duration]
        
        segment_output_path = os.path.join(output_folder, f"T{index}_{i}.wav")
        
        if len(segment) == segment_duration:
          segment.export(segment_output_path, format="wav")
           
          segments_df.loc[len(segments_df)] = [segment_output_path, label]

csv_path = "/content/drive/MyDrive/DB/file_path.csv"


output_folder = "/content/drive/MyDrive/DB/DBTotalSplitted"

# Check if the folder exists
os.makedirs(output_folder, exist_ok=True)

df = pd.read_csv(csv_path)

segments_df = pd.DataFrame(columns=["path", "label"])

for index, row in df.iterrows():
    wav_path = row["path"]
    label = row["label"]

    split_wav_with_labels(wav_path, label, output_folder, index, 8)

segments_df.to_csv("/content/drive/MyDrive/DB/splitted_total.csv", index=False)
```
